```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
f
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
title_dict = {'fontsize':15, 'fontweight':'bold', 'color':'#808080', 'family':'Verdana'}
label_dict = {'fontsize':12, 'color':'#808080', 'family':'Verdana'}
plt.title('График f(t)', fontdict=title_dict)
plt.xlabel('Значения t', fontdict=label_dict)
plt.ylabel('Значения f', fontdict=label_dict)
plt.plot(t, f, color='green')
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3 *x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title('График x**2')
ax2.plot(x, y2)
ax2.set_title('График 2x + 0.5')
ax3.plot(x, y3)
ax3.set_title('График -3x - 1.5')
ax4.plot(x, y4)
ax4.set_title('График sin(x)')
plt.subplots_adjust(wspace=0.3, hspace=0.3)
fig.set_size_inches(8, 6)
```
